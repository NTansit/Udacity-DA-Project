## Investigation Overview

In this investigation, I focus on factors that effect on ride duration.
I start by introducing ride duration distribution, followed by user type to see different distribution for each type.
Then plot realtionship between ride duration and usage hour classified by user type in clustered bar chart.

## Dataset Overview

This data set includes information about 174952 individual rides(after remove rides that included any null values) made in a bike-sharing system covering the greater San Francisco Bay area in February 2019. Provided information for each ride are ride duration, start/edn time and location,user type,member birth year,ender,bike share for all trip.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike_clean = pd.read_csv('bike_clean.csv')
```

## Distribution of Ride Duration

This plot shows distribution of ride duration from 0 to 75 minutes. Duration has a long-tailed distribution. Most of customers ride from 2 to 20 minutes. Peak at 5 minutes.

```python
binsize = 2
bins = np.arange(0, bike_clean['duration_min'].max()+binsize, binsize)

plt.figure(figsize=[15, 5])
plt.hist(data = bike_clean, x = 'duration_min', bins = bins)
plt.xlim([0,80])
plt.xticks(np.arange(0,80,5))
plt.xlabel('Duration (min)')
plt.title('Distribution of Ride Duration')
plt.show()
```

## Ride Duration vs. User types

This plot shows that casual customers ride longer than subscribers (13 > 8 mins compared by median). Ride duration of subscribers have less varient.

```python
plt.figure(figsize=[10, 6])
base_color = sb.color_palette()[0]
sb.boxplot(data=bike_clean, x='user_type', y='duration_min', color=base_color);
plt.title('Ride Duration vs. User types')
plt.yticks(np.arange(0,50,5))
plt.xlabel('User Types')
plt.ylabel('Ride Duration (Min)')
plt.ylim([0,50]);
```

## Ride duration vs. Usage Hour Classified by user type
Casual customers always ride longer than subscribers. Ride duration of subscribers are very stable (less fluctuate) compared to casual customers. Surprisingly ride duration peak at 3.00.

```python
plt.figure(figsize=[20, 6])
sb.barplot(data=bike_clean, x='start_hour', y='duration_min', ci=None, hue='user_type');
ticks = [00.00,01.00,02.00,03.00,04.00,05.00,06.00,07.00,08.00,09.00,10.00,11.00,12.00,13.00,14.00,15.00,16.00,17.00,18.00,19.00,20.00,21.00,22.00,23.00]
labels = ['00.00','01.00','02.00','03.00','04.00','05.00','06.00','07.00','08.00','09.00','10.00','11.00','12.00','13.00','14.00','15.00','16.00','17.00','18.00','19.00','20.00','21.00','22.00','23.00']
plt.xlabel('Usage Time');
plt.ylabel('Ride Duration (Min)')
plt.yticks(np.arange(0,90,5))
plt.title('Ride duration vs. Usage Hour Classified by user type')
plt.xticks(ticks, labels);
```
